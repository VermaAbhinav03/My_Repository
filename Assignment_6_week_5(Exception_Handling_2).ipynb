Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

LookupError class is used to handle errors related to lookup operations. A lookup operation typically involves searching for a value or key in a collection or mapping data structure.

The LookupError class serves as a base class for more specific lookup-related exception classes. It provides a way to catch and handle errors that occur during lookup operations without having to catch each specific error individually.

Here are a few specific cases where LookupError and its subclasses are commonly used in Python:

KeyError: This subclass of LookupError is raised when a dictionary key is not found. For example, if you try to access a dictionary with a key that doesn't exist, a KeyError is raised.

```python
import logging
logging.basicConfig(filename = "error.log" , level = logging.ERROR)
my_dict = {'apple': 1, 'banana': 2, 'orange': 3}
try:
    print(my_dict['grape'])
except KeyError as e:
    logging.exception("Q4.1. The key 'grape' does not exist in the dictionary {}".format(e))
```

IndexError: This subclass of LookupError is raised when accessing a sequence (such as a list or tuple) with an invalid index. If you try to access an element at an index that is out of range, an IndexError is raised.

```python
import logging
logging.basicConfig(filename = "error.log" , level = logging.ERROR)
my_list = [1, 2, 3, 4, 5]

try:
    print(my_list[10])
except IndexError as e :
    logging.exception("Q4.2. Invalid index. The list does not have an element at index 10{}".format(e))
```

Q5. Explain ImportError. What is ModuleNotFoundError?

ImportError: ImportError is a base exception class in Python that is raised when there is a problem with importing a module or an object from a module. It serves as a general error category for import-related issues. When an import statement fails, Python raises an ImportError with a specific subclass to provide more details about the specific import error. 
Example:--

```python
import logging
logging.basicConfig(filename = "error.log" , level = logging.ERROR)
try:
    import missing_module  # Incorrect import statement
except ImportError as e:
    logging.exception("Q5.1.Error importing module")
```

ModuleNotFoundError: ModuleNotFoundError is a specific subclass of ImportError that is raised when a module or package cannot be found during the import process. It is used when the Python interpreter is unable to locate the specified module or package.
Example:---

```python
import logging
logging.basicConfig(filename = "error.log" , level = logging.ERROR)
try:
    from non_existent_module import some_function  # Incorrect import statement
except ModuleNotFoundError as e:
    logging.exception("Q5.2.Module not found")
```

Q6. List down some best practices for exception handling in python.

Here are some best practices for exception handling in Python summarized:

1.Be specific: Catch only the exceptions you expect and avoid using a bare except statement.
2.Use multiple except blocks: Handle different types of exceptions separately.
3.Handle exceptions at the right level: Handle exceptions where you have enough context.
4.Use finally: Perform cleanup operations using a finally block.
5.Avoid overly broad exception handling: Only catch exceptions you can handle effectively.
6.Use exception chaining: Preserve the original exception traceback when re-raising exceptions.
7.Log exceptions: Use a logging framework to record exception details.
8.Consider performance impact: Optimize exception handling code if performance is a concern.
9.Test exception handling: Write test cases to ensure the error handling logic works as expected.
10.Follow PEP 8 guidelines: Adhere to the Python style guide for clean and readable code.

These practices will help you write effective and maintainable exception handling code in Python.
