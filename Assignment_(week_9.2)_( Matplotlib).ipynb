Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a powerful data visualization library in Python that allows you to create various types of plots and charts. Here’s a brief overview:

What is Matplotlib?
Matplotlib is a low-level library for data visualization in Python.
It is built on top of NumPy arrays and provides tools to create 2D plots of arrays.
Introduced by John Hunter in 2002, it emulates MATLAB-like graphs and visualizations.

Why is Matplotlib Used?
Visualization: Matplotlib enables us to visually explore and understand large amounts of data.
Customization: It offers extensive customization options for plots, allowing you to tailor them to your needs.
Integration: Matplotlib works seamlessly with other scientific libraries like SciPy.

Five Plots Using Pyplot Module:
Line Plot: Connects data points with lines, useful for showing trends or relationships.
Bar Plot: Displays bars representing data values, often used for comparing categories.
Histogram Plot: Shows the distribution of a continuous variable using bins.
Scatter Plot: Represents individual data points as dots, useful for identifying patterns.
Pie Charts: Displays proportions of different categories as slices of a pie.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

A scatter plot is a graphical representation that shows the relationship between two variables. It allows us to visualize how the values of one variable are related to the values of another. In a scatter plot:

The independent variable (or attribute) is plotted on the X-axis.
The dependent variable is plotted on the Y-axis.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data for x and y
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create the scatter plot
plt.scatter(x, y, color='c', marker='*', label='Data Points')
plt.title("Scatter Plot: Relationship between X and Y")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()
plt.legend()
plt.show()
```

The subplot() function in Matplotlib is used to create multiple plots within a single figure. It allows you to organize your plots in a grid layout, specifying the number of rows and columns. Each subplot can display different data or variations of the same data, making it easier to compare and analyze datasets.

Here’s how the subplot() function works:

You provide three arguments: the number of rows, the number of columns, and the index of the current plot.
The index determines the position of the current plot within the grid layout.
Now, let’s create four line plots using the subplot() function with the given data:

```python
import matplotlib.pyplot as plt
import numpy as np

# Example 1: 1x2 grid
x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplots(2, 2, figsize=(10, 8))

# First subplot (top left)
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title("line 1")
plt.grid(True)

# Second subplot (top middle)
plt.subplot(2, 2, 2)
plt.plot(x, y2, color="orange")
plt.title("line 2")
plt.grid(True)

# Third subplot (top right)
plt.subplot(2, 2, 3)
plt.plot(x, y3, color="green")
plt.title("line 3")
plt.grid(True)

# Fourth subplot (bottom left)
plt.subplot(2, 2, 4)
plt.plot(x, y4, color="red")
plt.title("line 4")
plt.grid(True)

plt.suptitle("four line plots")
plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

A box plot, also known as a box-and-whisker plot, is a graphical representation used to summarize the distribution of a dataset. It provides a concise view of key summary statistics, including the median, quartiles, and potential outliers. Here’s what you need to know about box plots:

Definition:
A box plot displays the following five summary statistics:
Minimum: The smallest value in the dataset.
First Quartile (Q1): The median of the lower half of the data.
Median (Q2): The middle value of the dataset, dividing it into two equal parts.
Third Quartile (Q3): The median of the upper half of the data.
Maximum: The largest value in the dataset.
It also indicates the interquartile range (IQR), which is the difference between Q3 and Q1.
Outliers (unusual observations) can be identified as individual points outside the whiskers.

Purpose and Use:
Box plots serve several purposes:
Distribution Summary: They summarize the distribution of data, especially when comparing multiple datasets.
Skewness Detection: They help identify whether the data distribution is skewed.
Outlier Detection: Outliers are visible as individual points beyond the whiskers.
Comparison: They allow easy comparison of distributions across different groups or datasets.
Box plots are particularly useful when dealing with large datasets or comparing multiple sets of data.

```python
import numpy as np
import matplotlib.pyplot as plt
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot([box1 , box2])
plt.show()
```
