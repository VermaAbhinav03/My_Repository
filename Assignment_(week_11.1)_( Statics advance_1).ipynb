Q3. Write a Python function to calculate the probability density function of a normal distribution with
given mean and standard deviation at a given point.

```python
import math

def normal_pdf(x, mean, std_dev):
   
    if std_dev <= 0:
        raise ValueError("Standard deviation must be positive.")

    exponent = -0.5 * ((x - mean) / std_dev) ** 2
    coefficient = 1 / (std_dev * math.sqrt(2 * math.pi))
    pdf_value = coefficient * math.exp(exponent)
    return pdf_value

# Example usage:
mean_value = 10.0
std_dev_value = 2.0
point_to_evaluate = 8.0
pdf_result = normal_pdf(point_to_evaluate, mean_value, std_dev_value)
print(f"PDF at x = {point_to_evaluate:.2f}: {pdf_result:.6f}")
```

Q4. What are the properties of Binomial distribution? Give two examples of events where binomial
distribution can be applied.

Properties of Binomial Distribution:
There are two possible outcomes: true or false, success or failure, yes or no.
There is a fixed number of (n) independent trials or a fixed number of (n) times repeated trials.
The probability of success or failure remains the same for each trial.
Examples of Events Where Binomial Distribution Can Be Applied: a. Tossing a Fair Coin:
Consider flipping a fair coin multiple times.
The outcome of each flip can be either heads (success) or tails (failure).
The binomial distribution can model the number of heads obtained in a fixed number of coin tosses.
b. Quality Control in Manufacturing:
Suppose a factory produces light bulbs, and each bulb can be either defective or non-defective.
The quality control team inspects a random sample of bulbs.
The binomial distribution can be used to analyze the number of defective bulbs in the sample.

Q5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4
and plot a histogram of the results using matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
sample_size = 1000
probability_success = 0.4

# Generate random sample from binomial distribution
random_sample = np.random.binomial(n=sample_size, p=probability_success, size=sample_size)

# Create a histogram
plt.figure(figsize=(8, 6))
plt.hist(random_sample, bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.title("Binomial Distribution (n=1000, p=0.4)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution
with given mean at a given point.

```python
import scipy.stats as stats

def poisson_cdf(point, mean):
    """
    Calculates the CDF of a Poisson distribution at a given point.

    Args:
        point (int): The value at which to evaluate the CDF.
        mean (float): Mean (average) of the Poisson distribution.

    Returns:
        float: The CDF value at the specified point.
    """
    if mean < 0:
        raise ValueError("Mean must be non-negative.")

    cdf_value = stats.poisson.cdf(point, mu=mean)
    return cdf_value

# Example usage:
mean_value = 3.5
point_to_evaluate = 2
cdf_result = poisson_cdf(point_to_evaluate, mean_value)
print(f"CDF at x = {point_to_evaluate}: {cdf_result:.6f}")
```

Explanation:

The poisson_cdf function takes two arguments: the value point at which to evaluate the CDF and the mean of the Poisson distribution.
It uses the scipy.stats.poisson.cdf function to compute the cumulative probability up to the specified point.

Q7. How Binomial distribution different from Poisson distribution?

While both the Binomial and Poisson distributions deal with random variables and probabilities, they differ in the nature of the experiment, number of trials, probability of success, and probability distribution.

Nature of the experiment: The Binomial distribution applies to experiments with a fixed number of trials, while the Poisson distribution applies to events occurring within a fixed interval.
Number of trials: The Binomial distribution requires a fixed number of trials, whereas the Poisson distribution does not have a predetermined number of events.
Probability of success: In the Binomial distribution, the probability of success remains constant for each trial, whereas the Poisson distribution does not explicitly consider the probability of success.
Probability distribution: The Binomial distribution represents the probability of achieving a specific number of successes, while the Poisson distribution models the probability of observing a certain number of events.

Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the
sample mean and variance.

```python
import numpy as np

# Parameters
sample_size = 1000
mean_value = 5

# Generate random sample from Poisson distribution
random_sample = np.random.poisson(lam=mean_value, size=sample_size)

# Calculate sample mean and variance
sample_mean = np.mean(random_sample)
sample_variance = np.var(random_sample)

print(f"Sample Mean: {sample_mean:.4f}")
print(f"Sample Variance: {sample_variance:.4f}")
```

Q9. How mean and variance are related in Binomial distribution and Poisson distribution?

Binomial Distribution:
Mean ((\mu)):
The mean of a binomial distribution is given by: (\mu = n \cdot p)
Here, (n) represents the number of trials, and (p) represents the probability of success in each trial.
Variance ((\sigma^2)):
The variance of a binomial distribution is given by: (\sigma^2 = n \cdot p \cdot (1-p))
The variance depends on both the number of trials and the probability of success.
Poisson Distribution:
Mean ((\lambda)):
The mean of a Poisson distribution is equal to its parameter (\lambda): (\mu = \lambda)
The parameter (\lambda) represents the average rate of events occurring in a fixed interval.
Variance ((\sigma^2)):
The variance of a Poisson distribution is also equal to its parameter (\lambda): (\sigma^2 = \lambda)
Unlike the binomial distribution, the variance is not affected by the probability of success; it solely depends on the average rate.
Comparison:
In summary:
Binomial: (n \cdot p > \sigma^2)
Poisson: (\lambda = \sigma^2)

Q10. In normal distribution with respect to mean position, where does the least frequent data appear?

Normal Distribution:
A normal distribution (also known as a Gaussian distribution or bell curve) is symmetrically shaped.
Most data values cluster around the central peak (the mean).
As we move away from the mean, the probabilities taper off equally in both directions.
Empirical Rule:
The empirical rule (also known as the 68-95-99.7 rule) provides insights into where most data values lie in a normal distribution:
Around 68% of values are within 1 standard deviation from the mean.
Around 95% of values are within 2 standard deviations from the mean.
Around 99.7% of values are within 3 standard deviations from the mean.
Tails Beyond the Mean:
The tails of the normal distribution represent extreme values.
Beyond 3 standard deviations from the mean, the data becomes increasingly rare.
The least frequent data points are found in these tails.
