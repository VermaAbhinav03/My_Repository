Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Scatter Plot: Visualizes the distribution of data points in a two-dimensional space. Useful for identifying patterns or correlations.
Line Plot: Shows the trend or change in a variable over time or another continuous dimension.
Pair Plot: Displays pairwise relationships between multiple variables in a dataset
Bar Plot: Compares the values of different categories using bars. Useful for comparing discrete data points.
Count Plot: Displays the count of occurrences for each category.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri = sns.load_dataset("fmri")

# Create a line plot for different events and regions
sns.set_theme(style="darkgrid")
sns.lineplot(x="timepoint", y="signal", hue="region", style="event", data=fmri )

# Display the plot
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic = sns.load_dataset("titanic")

# Create subplots for Age vs. Pclass and Fare vs. Pclass
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Box plot for Age vs. Pclass
sns.boxplot(x="pclass", y="age", data=titanic, ax=axes[0])
axes[0].set_title("Age vs. Pclass")
axes[0].set_xlabel("Passenger Class")
axes[0].set_ylabel("Age")

# Box plot for Fare vs. Pclass
sns.boxplot(x="pclass", y="fare", data=titanic, ax=axes[1])
axes[1].set_title("Fare vs. Pclass")
axes[1].set_xlabel("Passenger Class")
axes[1].set_ylabel("Fare")

plt.tight_layout()
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns

diamonds = sns.load_dataset("diamonds")
sns.histplot(data=diamonds, x="price", hue="cut", kde=True, palette="husl")
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris_data = sns.load_dataset("iris")

# Create a pair plot with different colors for each species
sns.pairplot(iris_data, hue="species", palette="husl")
```
